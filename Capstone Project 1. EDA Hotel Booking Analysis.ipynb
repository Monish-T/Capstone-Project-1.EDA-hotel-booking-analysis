### Variables Description 



1.   hotel : Type of Hotel (H1 = Resort Hotel or H2 = City Hotel).
2.   is_canceled : Describes wether the booking was canceled(1) or not(0). 
3.   lead_time : Number of days that elapsed between the entering date of the booking into the PMS and the arrival date
4.   arrival_date_year : Year of arrival date
5.   arrival_date_month : Month of arrival date
6.   arrival_date_week_number : Week number of year for arrival date
7.   arrival_date_day_of_month : Day of arrival date
8.   stays_in_weekend_nights : Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel
9.   stays_in_week_nights : Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel
10.  adults : Number of adults
11.  children : Number of children
12.  babies : Number of babies
13.  meal : Type of meal booked. Categories are presented in standard hospitality meal packages: Undefined/SC – no meal
14.  country : Country of origin. Categories are represented in the ISO 3155–3:2013 format
15.  market_segment : Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators”
16.  distribution_channel : Booking distribution channel.
17.  is_repeated_guest : Value indicating if the booking name was from a repeated guest (1) or not (0)
18.  previous_cancellations : Number of previous bookings that were cancelled by the customer prior to the current booking
19.  previous_bookings_not_canceled : Number of previous bookings not cancelled by the customer prior to the current booking
20.  reserved_room_type : Code of room type reserved.
21.  assigned_room_type : Code for the type of room assigned to the booking.
22.  booking_changes : Number of changes/amendments made to the booking from the moment the booking was entered on the PMS
23.  deposit_type : Indication on if the customer made a deposit to guarantee the booking. This variable can assume three categories: No Deposit – no deposit was made; Non Refund – a deposit was made in the value of the total stay cost; Refundable – a deposit was made with a value under the total cost of stay.
24.  agent : ID of the travel agency that made the booking
25.  company : ID of the company/entity that made the booking or responsible for paying the booking.
26.  days_in_waiting_list : Number of days the booking was in the waiting list before it was confirmed to the customer
27.  customer_type : Type of booking, assuming one of four categories: Contract - when the booking has an allotment or other type of contract associated to it; Group – when the booking is associated to a group; Transient – when the booking is not part of a group or contract, and is not associated to other transient booking; Transient-party – when the booking is transient, but is associated to at least other transient booking
28.  adr : Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights
29.  required_car_parking_spaces : Number of car parking spaces required by the customer
30.  total_of_special_requests : Number of special requests made by the customer
31.  reservation_status : Reservation last status, assuming one of three categories: Canceled – booking was canceled by the customer; Check-Out – customer has checked in but already departed; No-Show – customer did not check-in and did inform the hotel of the reason why
32.  reservation_status_date : Date at which the last status was set.

## 3. ***Data Wrangling***

### Data Wrangling Code

#### Initial Cleaning Of data for analysis

````python
# Write your code to make your dataset analysis ready.
# Creating a copy of dataset
df=dataset.copy()
````

````python
# Treating the Null values. 
# Replacing Null values in column agent and company with 0.
df['agent'].fillna('0',inplace=True)
df['company'].fillna('0',inplace = True)
df['children'].fillna('0',inplace=True)
# Replacing Null values in column country with Others.
df['country'].fillna('Others',inplace=True)
````

````python
# Changing the datatype of columns
# children, agent and company column are changed to int datatype
df['children'] = df['children'].astype(int)
df['agent'] = df['agent'].astype(int)
df['company'] = df['company'].astype(int)
# reservation_status_date column datatype changed to int
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
````

````python
# plotting box plot. To check if there are any outliers in the data.
# creating a list of numerical variable names.
numerical_var = df.describe().columns.to_list()
# iterating over column names to plot graph for each numerical column. 
for col in numerical_var:
  # setting size of graph.
  fig = plt.figure(figsize=(5,4))
  # plotting the graph
  sns.boxplot(x=df[col])
  # adding grid for better visualization
  plt.grid()
plt.show()
````

From the box plot we can derive that adr has a outlier value. lets try to find details about it.

````python
# checking the top 10 values of adr.
df['adr'].value_counts().sort_index(ascending=False).head(10)
````

````python
# Calculating the mean of adr column to compare with th outlier value.
df['adr'].mean()
````

````python
# we can conclude that 5400 in adr column is a outlier value as it is far from mean and also 10 times higher than 2nd largest value.
# Removing the outlier value. As, it is only as single value.
df = df[df['adr']!=5400]
````

````python
# creating a new column arrival date by combining the three year, month and day column
df.rename(columns={'arrival_date_year':'year','arrival_date_month':'month','arrival_date_day_of_month':'day'},inplace=True)
# creating a Dictionary on Month and corresponding month number.
list_index = list(df['month'].value_counts().index)
dict_month=dict(zip(list(df['month'].value_counts().index),[8,7,5,10,4,6,9,3,2,11,12,1]))
# using dictionary to change month name to month number
df['month']=df['month'].map(dict_month)
# combining three columns to create Arrival date column
df['arrival_date'] = pd.to_datetime(df[['year','month','day']])
````

````python
# Creating a new calculated column total stay. To find the total number of days each guest has booked. 
df['total_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
````

#### BI Variate and multi variate Analysis

````python
# categorical - Categorical Analysis
# difference in cancellation between city and resort hotel
# creating a new dataframe by grouping on the columns
hc_df = df.groupby(['hotel'])['is_canceled'].value_counts().unstack().rename(columns={0:'not_cancelled',
                                                                                      1:'cancelled'})       
# Calculating cancellation percentage.
hc_df['cancellation_percent'] = hc_df['cancelled'] * 100 /(hc_df['not_cancelled'] +hc_df['cancelled'])
hc_df
````

Total of 41% city Hotel booking are cancelled, which is very high compared to Resort Hotels where the cancellation percentage is only 27%

````python
# Numerical and categorical variables
# Hotel bookings acc to hotel type vs lead_time
# creating a new dataframe by grouping on the columns
hot_lead_df = df.groupby(['lead_time'])['hotel'].value_counts().unstack().head(30)
hot_lead_df['Booking_percent_difference'] = hot_lead_df['City Hotel'] * 100 /hot_lead_df['Resort Hotel']          # No significant. and also further use
hot_lead_df
````

````python
# Numerical and categorical variables
# cancellation vs lead_time
# creating a new dataframe by grouping on the columns
can_lead_df = df.groupby(['lead_time'])['is_canceled'].value_counts().unstack().rename(columns={0:'not_cancelled',
                                                                                      1:'cancelled'})
# calculating cancellation percentage
can_lead_df['cancellation_percent'] = can_lead_df['cancelled'] * 100 /(can_lead_df['not_cancelled']  + can_lead_df['cancelled'])       
can_lead_df.head(20)
````

As the lead time increases the cancellation percentage also increases. It increases from 10% to 30%.

````python
# Multi Variate 
# Comparing the cancellation count with increasing lead time for both the hotel types
# creating a new dataframe by grouping on the columns
hot_can_lead_df = df.groupby(['lead_time','hotel'])['is_canceled'].value_counts().unstack().rename(columns={0:'not_cancelled',
                                                                                                            1:'cancelled'})
# calculating cancellation percentage
hot_can_lead_df['cancellation_percent'] = hot_can_lead_df['cancelled'] * 100 /(hot_can_lead_df['not_cancelled']  + hot_can_lead_df['cancelled'])
hot_can_lead_df.unstack().head(30)
````

As the lead time increases the cancellation percentage also increases for both the hotel types. It increases from 5% to 30% for Resort Hotels and for city hotel it goes from 10% to 45%..
